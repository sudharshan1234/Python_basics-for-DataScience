```
"""
you are a teacher and you have different student records containing id of  student and marks list in each subj where diff students have taken different number of subjects
all these records are in hard copy and we shud enter it to computer and want to compute the avg marks of each student and display
"""
```

```
def getData():
  D={}
  while True:
    id=input("Enter student id: ")
    marksList=input("Enter the marks by comma separated values:")
    moreStudents=input("more? Yes/No: ")
    if id in D:
      print(id," is already present")
    else:
      D[id]=marksList.split(",")
    if(moreStudents.lower()=="no"):
      return D
```

```
studentData=getData()
```

```
studentData
```

```
def getAvg(D):
  avg={}
  for x in D:
    L=D[x]
    sum=0
    for i in L:
      sum+=int(i)
    avg[x]=sum/len(L)
  return avg    
```

```
avgM=getAvg(studentData)
```

```
for x in avgM:
  print('Student:',x,'Marks:',avgM[x])
```
